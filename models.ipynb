## **Load Data**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/AashrithC/AP-Research/main/data.csv')
df
```

##Data Preparation

###Data Separation as X and Y

```python
y = df['SelfRepHapp']
y
```

```python
X = df.drop('SelfRepHapp', axis=1)
X
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, shuffle=True)
```

```python
X_train
```

```python
X_test
```

## **Model Building**

### Linear Regression

####Training the Model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

####Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

####Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

### Random Forest

####Training the Model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

#### Applying the model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

####Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

### MLP

####Training the Model

```python
from sklearn.neural_network import MLPRegressor

regr = MLPRegressor(random_state=1, max_iter=100)
regr.fit(X_train, y_train)
```

#### Applying the model to make a prediction

```python
y_regr_train_pred = regr.predict(X_train)
y_regr_test_pred = regr.predict(X_test)
```

####Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

regr_train_mse = mean_squared_error(y_train, y_regr_train_pred)
regr_train_r2 = r2_score(y_train, y_regr_train_pred)

regr_test_mse = mean_squared_error(y_test, y_regr_test_pred)
regr_test_r2 = r2_score(y_test, y_regr_test_pred)
```

```python
regr_results = pd.DataFrame(['MLP', regr_train_mse, regr_train_r2, regr_test_mse, regr_test_r2]).transpose()
regr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
regr_results
```

###Model Comparison

```python
df_models = pd.concat([lr_results, rf_results, regr_results], axis=0)
df_models
df_models.reset_index(drop=True)
```
